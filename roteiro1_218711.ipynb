Import das bibliotecas a serem utilizadas no roteiro.

```
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
from scipy.io import wavfile
from scipy.fft import fft, fftfreq
```

Agora carregamos o arquivo de áudio dado no roteiro com o comando wavfile.read:


```
fs, som = wavfile.read('elefante.wav')
ipd.Audio(som,rate=fs)
```

Reproduzindo os prints do roteiro original:

```
t = np.arange(som.size)/fs
plt.plot(t,som);
plt.title('Gráfico do som do elefante')
plt.xlabel('t (s)');
```

Agora aproximando o gráfico e forçando o pyton a plotar os pontos específicos do vetor `som`.

```
t = np.arange(som.size)/fs
plt.plot(t,som);
plt.plot(t,som,'ro',fillstyle = 'none')
plt.title('Gráfico do som do elefante')
plt.xlabel('t (s)');
plt.xlim((1,1.005));
```

Agora, devemos gerar o espectro do sinal e plotar o gráfico, com as frequências analógicas no eixo x:

Pela relação da DFTFT sabemos que $\Omega = 2\pi k * \frac{1}{N} \rightarrow N = \frac{2\pi k}{\Omega}$

```
y = np.fft.fft(som)

N = len(y) + 1
y = np.abs(y)

freq = np.linspace(0,fs,y.size)
print(freq)
print("Taxa de amostragem é dada por:",fs)

plt.figure(figsize=(8,4))
plt.title("Transformada de Fourrier do sinal dado")
plt.xlabel("Frequência $\Omega$ [Hz]")
plt.ylabel("Valor da Transformada")
plt.plot(freq,y)

```
